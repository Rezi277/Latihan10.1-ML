```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import math
```

```python
df1 = {
    'temperatur': [10, 25, 15, 20, 18, 20, 22, 24],
    'kecepatan_angin': [0, 0, 5, 3, 7, 10, 5, 6],
    'klasifikasi': ['Dingin', 'Panas', 'Dingin', 'Panas', 'Dingin', 'Dingin', 'Panas', 'Panas']
}
```

```python
df1 = pd.DataFrame(df1)
```

```python
x= df1[['temperatur', 'kecepatan_angin']]
y= df1['klasifikasi']
```

```python
replacements = {
    'klasifikasi': {'Dingin': 0, 'Panas': 1}
}
df1 = df1.replace(replacements, inplace=False)
df1.head()
```

```python
test_point = {'temp': 16, 'angin': 3}
```

```python
def hitung_jarak(data1, data2):
    distance = math.sqrt((data1['temp'] - data2['temp'])**2 + (data1['angin'] - data2['angin'])**2)
    return distance
```

```python
neighbors = []
for index, row_data in df1.iterrows():
    # Prepare the row data in a dictionary format compatible with hitung_jarak
    current_row_for_distance = {
        'temp': row_data['temperatur'],
        'angin': row_data['kecepatan_angin']
    }
    jarak = hitung_jarak(test_point, current_row_for_distance)
    neighbors.append({
        'jarak': jarak,
        'kelas': row_data['klasifikasi'], # Access 'klasifikasi' from the Series
        'df1': f"Temp: {row_data['temperatur']}, Angin: {row_data['kecepatan_angin']}" # Use actual column names
    })
    print(f"Jarak ke ({row_data['temperatur']}, {row_data['kecepatan_angin']}) [{row_data['klasifikasi']}] = {jarak:.2f}")
```

```python
neighbors.sort(key=lambda x: x['jarak'])
k = 3
k_nearest = neighbors[:k]

print(f"\n--- {k} Tetangga Terdekat ---")
votes = []
for neighbor in k_nearest:
    print(f"Jarak: {neighbor['jarak']:.2f} -> Kelas: {neighbor['kelas']}")
    votes.append(neighbor['kelas'])
```

```python
hasil_prediksi = max(set(votes), key=votes.count)

# Map the numerical prediction back to its original string label
# Assuming 0 is 'Dingin' and 1 is 'Panas' based on previous replacements
if hasil_prediksi == 0:
    persepsi_marry_label = 'Dingin'
elif hasil_prediksi == 1:
    persepsi_marry_label = 'Panas'
else:
    persepsi_marry_label = str(hasil_prediksi) # Fallback if unexpected value

print(f"\n--- Hasil Akhir ---")
print(f"Berdasarkan Voting {k} tetangga terdekat, Persepsi Marry adalah: {persepsi_marry_label.upper()}")
```
